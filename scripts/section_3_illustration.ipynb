```python
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt
import matplotlib.figure as pltfig
from sklearn.linear_model import LinearRegression

l = 1
u = 3
def f(x): return (x - 1.75)**2
N = 1
D = np.array([l, 1.75, 2.25, u])
D = D.reshape(np.size(D), 1)

X = (l - 0.2) + np.arange(0, 1, 0.01) * ((u + 0.2) - (l - 0.1))
plt.plot(X, f(X), '--', color = 'black')

my_dpi = 300

# plt.savefig('../results/figures/parabola01.png', dpi = my_dpi)

plt.scatter(x = D, y = f(D), color = 'black')

# plt.savefig('../results/figures/parabola02.png', dpi = my_dpi)

reg = LinearRegression().fit(D, f(D))
# plt.plot(D, reg.predict(D), linewidth = 3)
tmpX = X.reshape(np.size(X), 1)
plt.plot(tmpX, reg.predict(tmpX), linewidth = 3)

# plt.savefig('../results/figures/parabola03.png', dpi = my_dpi)

y_limits = plt.ylim()
plt.plot(D, (y_limits[0] + 0.013) * np.ones(D.size), linewidth = 3, color = 'orangered')
plt.ylim(y_limits)

# plt.savefig('../results/figures/parabola04.png', dpi = my_dpi)

tmp = np.array([1])
tmp = tmp.reshape(np.size(tmp), 1)
plt.plot(tmp, reg.predict(tmp), color = 'red', marker = '*', markersize = 20)

# plt.savefig('../results/figures/parabola05.png', dpi = my_dpi)

plt.fill(D, f(D), facecolor = 'lightsalmon', edgecolor = 'orangered', linewidth = 3)

# plt.savefig('../results/figures/parabola06.png', dpi = my_dpi)

tmp = np.array([1.37])
tmp = tmp.reshape(np.size(tmp), 1)
plt.plot(tmp, reg.predict(tmp), color = 'darkgreen', marker = '*', markersize = 20)

# plt.savefig('../results/figures/parabola07.png', dpi = my_dpi)

plt.show()
```

```python
w, h = pltfig.figaspect(1/2.5)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(w, h))

ax1.plot(X, f(X), '--', color = 'black')
ax2.plot(X, f(X), '--', color = 'black')

ax1.scatter(x = D, y = f(D), color = 'black')
ax2.scatter(x = D, y = f(D), color = 'black')

ax1.plot(tmpX, reg.predict(tmpX), linewidth = 3, color = 'gray')
ax2.plot(tmpX, reg.predict(tmpX), linewidth = 3, color = 'gray')

y_limits = ax1.get_ylim()
ax1.plot(D, (y_limits[0] + 0.013) * np.ones(D.size), linewidth = 3, color = 'gray')
ax1.set_ylim(y_limits)

tmp = np.array([1])
tmp = tmp.reshape(np.size(tmp), 1)
ax1.plot(tmp, reg.predict(tmp), color = 'black', marker = '*', markersize = 20)

ax2.fill(D, f(D), facecolor = 'silver', edgecolor = 'dimgray', linewidth = 3)

tmp = np.array([1.37])
tmp = tmp.reshape(np.size(tmp), 1)
plt.plot(tmp, reg.predict(tmp), color = 'black', marker = '*', markersize = 20)

plt.savefig('../results/figures/section_3_illustration.png', dpi = my_dpi, bbox_inches='tight')

plt.show()
```

```python
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt
import matplotlib.figure as pltfig
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

l = 1
u = 3
def f(x): return (x - 1.75)**2
N = 1
D = np.array([l, l + 0.2, u])
D = D.reshape(np.size(D), 1)

X = (l - 0.2) + np.arange(0, 1, 0.01) * ((u + 0.2) - (l - 0.1))
plt.plot(X, f(X), '--', color = 'black')

my_dpi = 300

plt.scatter(x = D, y = f(D), color = 'black')

poly = PolynomialFeatures(degree = 2, include_bias = False)
poly_features = poly.fit_transform(D.reshape(-1, 1))
reg = LinearRegression().fit(poly_features, f(D))

# plt.savefig('../results/figures/parabola08.png', dpi = my_dpi)

# X = l + np.arange(0, 1, 0.01) * (u - l)
X = X.reshape(np.size(X), 1)
tmp = poly.fit_transform(X.reshape(-1, 1))
plt.plot(X, reg.predict(tmp), linewidth = 3)

y_limits = plt.ylim()
plt.plot(D, (y_limits[0] + 0.013) * np.ones(D.size), linewidth = 3, color = 'orangered')
plt.ylim(y_limits)

X = np.array([1.75])
X = X.reshape(np.size(X), 1)
tmp = poly.fit_transform(X.reshape(-1, 1))
plt.plot(X, reg.predict(tmp), color = 'red', marker = '*', markersize = 20)

#plt.savefig('../results/figures/parabola09.png', dpi = my_dpi)

plt.fill(D, f(D), facecolor = 'lightsalmon', edgecolor = 'orangered', linewidth = 3)

X = np.array([l + 0.2])
X = X.reshape(np.size(X), 1)
tmp = poly.fit_transform(X.reshape(-1, 1))
plt.plot(X, reg.predict(tmp), color = 'darkgreen', marker = '*', markersize = 20)

# plt.savefig('../results/figures/parabola10.png', dpi = my_dpi)

plt.show()
```
